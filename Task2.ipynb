```python
import csv 

def mean(lists):
    input_list = list(map(int,lists))
    print(input_list)
    total_mean = sum(input_list) / len(input_list)
    return round(total_mean,2) 

with open("yeoviltondata.csv",newline="",mode="r") as csvfile:
    data = csv.reader(csvfile)

    with open('yeoviltondata_yearly.txt', 'a', newline='') as csvfile:
        writerows = csv.writer(csvfile, delimiter=' ',quotechar='|', quoting=csv.QUOTE_MINIMAL)
        year = 1964
        rain_list = []
        t_max_list = []
        t_min_list = []
        writerows.writerow(["Year","T_Max","T_Min","Rain"])
        for i in data:
            if int(i[1]) == year:
                rain_list.append(i[2])
                t_max_list.append(i[3])
                t_min_list.append(i[4])
            else:
                writerows.writerow([str(year),str(mean(t_max_list)),str(mean(t_min_list)),str(mean(rain_list))])
                year += 1
            print("Year: "+str(i[1]))
            print("Rain: "+str(i[2]))
            print("T_Max: "+str(i[3]))
            print("T_Min: "+str(i[4]))
            print(i[1:5])
            #writerows.writerow(i[1:5])
```
